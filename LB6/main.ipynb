```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


df = pd.read_csv('alco.csv')
print(df.head())
```

### Построить любой график (любого рода) с помощью чистого matplotlib / pylab

```python
plt.hist(df['total_litres_of_pure_alcohol'], bins=20, alpha=0.5, color='b')
plt.xlabel('total_litres_of_pure_alcohol')
plt.ylabel('wine_servings')
plt.title('Distribution of Car Years')
plt.show()
```

### Построить несколько суб-графиков на одном графике используя .subplot.

```python
# построение точечной диаграммы и гистограммы
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# график рассеяния

ax1.scatter(df['spirit_servings'],df['wine_servings'], alpha=0.5, color='r')
ax1.set_xlabel('spirit_servings')
ax1.set_ylabel('wine_servings')
ax1.set_title('spirit_servings vs Popularity')

# histogram
ax2.hist(df['spirit_servings'], bins=20, alpha=0.5, color='b')
ax2.set_xlabel('spirit_servings')
ax2.set_ylabel('total_litres_of_pure_alcohol')
ax2.set_title('Distribution of Engine HP')

plt.show()
```

### На основе Вашего датасета построить три разных вида графиков используя pandas.DataFrame.plot / Series.plot

```python

df['beer_servings'].plot(kind='density', figsize=(8,5))
plt.xlabel('country')
plt.title('Beer Servings by Country')
plt.show()
```

### График разброса (scatter plot) - отображает точки на плоскости, где каждая точка представляет две числовые переменны

```python
df.plot.scatter(x='beer_servings', y='wine_servings', figsize=(8, 5))
plt.xlabel('Beer Servings')
plt.ylabel('Wine Servings')
plt.title('Beer Servings vs Wine Servings')
plt.show()
```

### Boxplot (ящик с усами) - отображает показатели центральной тенденции, размах и выбросы

```python
# Boxplot для столбца 'continent' и 'beer_servings'
df.boxplot(column='beer_servings', by='continent', figsize=(8, 5))
plt.xlabel('Continent')
plt.ylabel('Beer Servings')
plt.title('Beer Servings by Continent')
plt.suptitle('')  # Удаление заголовка, сгенерированного методом boxplot
plt.show()
```

### Построить .pairplot с помощью seaborn на каком-либо наборе данных, проанализировав взаимосвязь набора числовых переменных

```python
sns.pairplot(data=df, vars=['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol'])
```

### Построить .jointplot, проанализировав взаимосвязь двух конкретных числовых переменных

```python
sns.jointplot(data=df, x='beer_servings', y='spirit_servings', kind='reg')
```

### Построить .boxplot или .violinplot на каком-либо наборе данных, проанализировав какой-то числовой показатель в разрезе категориального

```python
sns.boxplot(data=df, x='beer_servings', y='continent')
```

### Построить график heatmap попарной корреляции всех числовых признаков (категориальные признаки убираете из графика)

```python
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
```

### Построить любой график (связанный с вашими данными) с помощью plotly 

```python
import plotly.express as px

# Построение графика связанного с данными с помощью Plotly
fig = px.bar(df, x='country', y='beer_servings', color='continent', title='Beer Servings by Country')
fig.show()
```
